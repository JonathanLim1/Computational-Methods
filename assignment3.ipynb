```python


x1 = np.array([-1+2/n1*i for i in range(0,n1+1)])
x2 = np.array([-1+2/n2*i for i in range(0,n2+1)])

A1 = np.vander(x1,m1+1,increasing=True)
A2 = np.vander(x2,m2+1,increasing=True)

y1 = np.array([1/(1+25*(x**2)) for x in x1])
y2 = np.array([1/(1+25*(x**2)) for x in x2])
```

Perform SVD on the Vandermonde Matrix A

```python
U1,s1,V1h = np.linalg.svd(A1,full_matrices=True)
U2,s2,V2h = np.linalg.svd(A2,full_matrices=True)
```

Least Square solution using SVD

```python
S1 = np.zeros((n1+1,m1+1))
S1[:m1+1,:m1+1] = np.diag(s1)
p1 = V1h.T @ np.linalg.inv(S1) @ U1.T @ y1

S2 = np.zeros((n2+1,m2+1))
S2[:m2+1,:m2+1] = np.diag(s2)
p2 = V2h.T @ np.linalg.pinv(S2) @ U2.T @ y2
```

# Plot 

```python
import matplotlib.pyplot as plt
x_plot = np.linspace(-1,1,1000)
y_plot = 1/(1+25*(x_plot**2))

def polynom(coeffs,x):
    ans = 0
    for i in range(len(coeffs)):
        ans += coeffs[i]*x**i
    return ans


y_1 = polynom(p1,x1)
y_2 = polynom(p2,x2)

plt.figure(figsize=(17,9))

plt.plot(x_plot,y_plot,label = 'Target function')
plt.plot(x1,y_1,label = "SVD Least Square (m=n=10)")
plt.plot(x2,y_2,label = "SVD Least Square (m=20,n=80)")
plt.legend()
plt.title('Target Function and least square polynomials')
plt.show()
```
